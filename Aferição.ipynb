como resultado um tuplo composto por duas listas, a primeira contendo os números menores ou iguais ao
número passado como argumento, e a segunda contendo os números maiores que o argumento.


```python
def splitLista(lista,numero):
    menoresouiguais=[]
    maiores=[]
    for elem in lista:
        if elem <= numero:
            menoresouiguais.append(elem)
        else: #elem maior que o número 
            maiores.append(elem)
    return (menoresouiguais,maiores)
    
print(splitLista([1,2,3,4,5], 3))
```

que n é o argumento da função.


```python
def multiplos23(n):
    multiplos=0
    i=1
    while i < n: 
        if i %23 == 0:
            multiplos = multiplos + 1
        i=i+1
    return multiplos 

print(multiplos23(100))
```
